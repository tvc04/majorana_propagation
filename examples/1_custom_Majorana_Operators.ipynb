Let's discuss how to create a `MajoranSum` with any Majorana string other than the ones currently supported by the constructors `MajroanaSum(nsites, :fermionic_symbol, sites)` for the most common fermionic operators.
Majorana propagation can handle any Majorana string 
\begin{equation}\mu(v) \equiv(i)^{v^T \omega_{\mathrm{L}} v} \cdot \gamma_1^{v_1} {\gamma_1^\prime}^{v_2} \cdots \gamma_N^{v_{2 N-1}} \gamma_N^{\prime v_{2 N}}\end{equation}

In the cell below we illustrate how to use a `MajoranaSum` constructors that accepts a list of integers denoting which Majorana operators are activated, based on the indexing (1). So for example for the string
$$\mu(v_1) \equiv \gamma_1 \gamma_1^\prime \gamma_3 \gamma_6^\prime$$
(notice that since $v^T \omega_{\mathrm{L}} v = (w(v)^2-w(v))/2~(\mathrm{mod}2)$ there is no prefactor for a string of weight $w(v_1)=4$) can be initialized by passing the vector `[1, 2, 5, 12]`, as the mapping from operators to indices is $$\begin{split}\gamma_j&\to 2j-1\\\gamma^\prime_j&\to 2j\end{split}$$

Similarly, the string 
$$\mu(v_2) = \gamma_1 \gamma_2^\prime \gamma_3 \gamma_4^\prime$$
can be generated by passing the indices `[1, 4, 5, 8]`.

```julia
n_fermions = 8 
is_spinful = false

v1 = [1, 2, 5, 12]
msum1 = MajoranaSum(n_fermions, v1, is_spinful)

v2 = [1, 4, 5, 8]
msum2 = MajoranaSum(n_fermions, v2, is_spinful)
```

After this, Majorana sums can be manipolated as usual, so for example their product is

```julia
msum = msum1 * msum2
```

## Spinful case

For the case of spinful fermions, the code stores the operators according to the following convention
$$\mu(v) \equiv(i)^{v^T \omega_{\mathrm{L}} v} \cdot \gamma_{1,\uparrow}^{v_1} {\gamma_{1,\uparrow}^\prime}^{v_2}\gamma_{1,\downarrow}^{v_3} {\gamma_{1,\downarrow}^\prime}^{v_4} \cdots \gamma_{N,\uparrow}^{v_{4N-3}} {\gamma_{N,\uparrow}^\prime}^{v_{4N-2}}\gamma_{N,\downarrow}^{v_{4N-1}} {\gamma_{N,\downarrow}^\prime}^{v_{4N}},$$
hence the mapping from operators to indices is $$\begin{split}\gamma_{j,\uparrow}&\to 4j-3\\\gamma^\prime_{j,\uparrow}&\to 4j-2\\\gamma_{j,\downarrow}&\to 4j-1\\\gamma^\prime_{j,\downarrow}&\to 4j\end{split}$$
Hence the string 
$$\mu(v_3) = \gamma_{1,\uparrow} \gamma_{2,\uparrow} \gamma_{3,\downarrow}^\prime \gamma_{4,\uparrow}^\prime$$
is implemented as

```julia
n_spinful_sites = 5
is_spinful = true
gammas = [1, 5, 12, 14]

msum = MajoranaSum(n_spinful_sites, gammas, is_spinful)
```
